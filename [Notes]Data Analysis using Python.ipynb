```python
# Now in our last function, if there is no value in any line, it will going to give us an error, for that we have to make some changes in our function
# To not get the value error, we will use if and else functions

def parse_values(data_line):
    values=[]
    for item in data_line.strip().split(','):
        if item =='':
            values.append(0.0)
        else:
            values.append(float(item))
    return values
```

```python
# Now lets create a function which take both list of headers and list of values to create a dictonary
# here I have used zip to combine 2 dictonries (values and headers)
def create_item_dict(values, headers):
    result={}
    for values, header in zip(values, headers):
        result[header]= values
    return result
```

```python
# use of zip by an Example 
for item in zip ([1,2,3], ['a','b','c']):
    print(item)
```

```python
def read_csv(path):
    result=[]
    # to open a file in read mode-
    with open(path, 'r') as f:
        # to get a list of lines
        lines=f.readlines()
        # Parse the headers
        headers=parse_headers(lines[0])
        # Loop over the remaining lines
        for data_line in lines[1:]:
            # Parse the values
            values=parse_values(data_line)
            # create a dic using values and headers
            item_dict = create_item_dict(values, headers)
            # Add the dictionary to the result
            result.append(item_dict)
    return result
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# WE used plt.plot to plot our graph
plt.plot(years, apples)
plt.plot(years, oranges)
# We used plt,xlabe1 and ylabe1 to lable x and y axis
plt.xlabel('Year')
plt.ylabel('Yield(tons per hectar)')
# We used plt.title to give the titel to our chart and .legend to define which line represent which colour
plt.title('Crop yield in kanto')
plt.legend(['Apples', 'Oranges']);
```

```python
plt.plot(years, apples)
plt.plot(years, oranges)

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Oranges']);
```

```python
# to change the background of our charts
sns.set_style('darkgrid')
```

```python
plt.plot(years, apples, 's-b')
plt.plot(years, oranges, 'o--r')

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Oranges']);
```

## 2. Scatter Plot 

```python
# lets import iris data set 
# load data into a pandas dataframe
flowers_df= sns.load_dataset("iris")
```

```python
flowers_df
```

```python
flowers_df.species.unique()
```

```python
# We have used "sns.scatterplot" to plot our graph
# we used "s", to inc/dec the size of dots
# we used plt.figure and plt.title to give our chart a title and a size
plt.figure(figsize=(12,6))
plt.title('Sepal Dimensions')
sns.scatterplot(x=flowers_df.sepal_length,
                y=flowers_df.sepal_width, 
                hue=flowers_df.species,
                s=50);
```

## 3. Histogram 

```python
# Loasd data into a pandas dataframe
flowers_df= sns.load_dataset("iris")
```

```python
flowers_df.sepal_width
```

```python
plt.hist(flowers_df.sepal_width);
```

```python
# if we want to change the bin size
import numpy as np
plt.title('Distribution of sepal width')
plt.hist(flowers_df.sepal_width, bins=np.arange(2,5, 0.25));
```

```python
# TO PLOT MULTIPLE HISTOGRAM
setosa_df=flowers_df[flowers_df.species=='setosa']
versicolor_df=flowers_df[flowers_df.species=='versicolor']
verginica_df=flowers_df[flowers_df.species=='virginica']
```

```python
plt.hist(setosa_df.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25));
plt.hist(versicolor_df.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25));
plt.title('Distribution of sepal width')
plt.legend(['Setosa', 'Versicolor']);
```

```python
# to stake multiple histogram on top of one another
plt.title('Distribution of Sepal Width')
plt.hist([setosa_df.sepal_width,
         versicolor_df.sepal_width,
         verginica_df.sepal_width],
        bins=np.arange(2,5,0.25),
        stacked=True);
plt.legend(['Setosa', 'Versicolor', 'Virginica']);
```

## If we want to plot multiple charts in a Grid 

```python
fig, axes=plt.subplots(2,3, figsize=(16,8))

# use the axes for plotting
axes[0,0].plot(years, apples, 's-b')
axes[0,0].plot(years, oranges, 'o--r')
axes[0,0].set_xlabel('Year')
axes[0,0].set_ylabel('Yield (tons per hectare)')
axes[0,0].set_title("Crop Yields in Kanto")
axes[0,0].legend(['Apples', 'Oranges']);

# Pass the axes into seaborn
axes[0,1].set_title('Sempal Length vs. sepal width')
sns.scatterplot(x=flowers_df.sepal_length,
               y=flowers_df.sepal_width,
               hue=flowers_df.species,
               s=50,
               ax=axes[0,1]);

# use the axes for plotting
axes[0,2].set_title('Distribution of Sepal Width')
axes[0,2].hist([setosa_df.sepal_width,
         versicolor_df.sepal_width,
         verginica_df.sepal_width],
        bins=np.arange(2,5,0.25),
        stacked=True);

axes[0,2].legend(['Setosa', 'Versicolor', 'Virginica']);

# use the axes into seaborn
axes[1,0].set_title('Resturant Bills')
sns.barplot(x='day', y='total_bill', hue='sex', data=tips_df, ax=axes[1,0]);

# Pass the axes into seaborn
axes[1,1].set_title('Flight traffic')
sns.heatmap(flights_df, fmt='d', annot=True, cmap='Blues', ax=axes[1,1]);

# Plot an image using the axes
axes[1,2].grid(False)
axes[1,2].set_title('Just For Fun')
axes[1,2].axis('off')
axes[1,2].imshow(img);

plt.tight_layout(pad=2);
```
